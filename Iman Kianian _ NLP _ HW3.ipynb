```python
def Viterbi(test_obser, A=A, B=B, train = train,tags = train_tags):
    states = []
    Tags = list(tags) # List of Unique Tags
    
    for i, word in enumerate(test_obser):
        # list of probability for each word
        p_list = [] 
        for tag in Tags:
            if i == 0:
                transition = A.loc['DELM', tag]
            else:
                transition = A.loc[states[-1], tag]        
            # get emission and state probabilities from B matrix
            emission =  B.loc[word,tag] #P_Wi_Ti(word, tag)
            state = emission * transition    
            p_list.append(state)  
        max_p = max(p_list)
        # get maximum probability state 
        max_state = Tags[p_list.index(max_p)] 
        states.append(max_state)
        
        if (i+1) % 10000 == 0 :
            print(i+1,end='\r')
    return list(zip(test_obser, states))
```

```python
tagged_seq = Viterbi(test_untagged)
```

```python
y_pred = [t[1] for t in tagged_seq]
```
